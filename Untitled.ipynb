```python
df = pd.read_csv(csv_file)
```

```python
mask = (df['Team'] == 'India') & (df['match_no'] == 49)
```

```python
final_india = df[mask]
```

```python
final_india.groupby(['batsmen','runs']).value_counts()
```

```python
indian_batting = final_india.groupby(['batsmen']).agg({"extras": 'sum',"runs" : 'sum'})
```

```python
indian_batting.loc['Total'] = [indian_batting['extras'].sum(),indian_batting['runs'].sum()]
```
